```python
for root,dirs,files in os.walk(path):
    print("Read datas from the directory : ", root)
    xlsfiles=[ _ for _ in files if _.endswith('.xls') ]
    for xlsfile, index in zip(xlsfiles, range(0, len(xlsfiles))) :
        print(index, root + "/" + xlsfile)
        tutoring = pd.read_excel(root + "/" + xlsfile, index_col=0)
```

```python
import re
email = tutoring['Email']
cleaned_email = []
for _ in email:
    email.dropna(inplace=True)
    cleaned_email = np.append(cleaned_email , re.search('@([\w+\-]*)', _).group(1))

unique, counts = np.unique(cleaned_email, return_counts=True)
df = pd.DataFrame(data={"Email":unique, "Counts":counts}).set_index('Email')
df
```
